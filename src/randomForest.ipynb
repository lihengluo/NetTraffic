```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys

```

```python
# 使用随机森林算法实现分类
# 其中labels 和 features 分别是标签和特征
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 读取数据， 第一行是列名
labels = pd.read_csv('../data/labels.csv', header=0)
featuress = pd.read_csv('../data/features.csv', header=0)
# 将数据转换为numpy数组
labels = np.array(labels)
featuress = np.array(featuress)
# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(featuress, labels, test_size=0.3, random_state=0)

#加入网格调参
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [1, 2, 3, 4, 5]
}
# 训练模型,使用GPU训练
clf = GridSearchCV(RandomForestClassifier(random_state=0), param_grid, cv=5, n_jobs=1)
clf.fit(X_train, y_train)

# 输出最优参数
print(clf.best_params_)

# 预测
y_pred = clf.predict(X_test)

# 评估
print(accuracy_score(y_test, y_pred))

# 保存模型
import pickle
with open('../model/rf.pkl', 'wb') as f:
    pickle.dump(clf, f)
```
